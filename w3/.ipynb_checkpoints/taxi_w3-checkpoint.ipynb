В рамках моделей ARIMA можно учесть только одну из них. Обычно в таких случаях сезонность с самым маленьким периодом явно моделируют с помощью аримы, а все остальные учитывают за счёт регрессионной компоненты.

Для учёта недельной сезонности используются регрессионные признаки следующего вида:

$s_i = \sin \left( [1,\dots,T] * 2 \pi i/ 168 \right), c_i = \cos \left( [1,\dots,T] * 2 \pi i/ 168 \right), i=1,\dots, K.$

Здесь T — это длина моделируемого ряда, 168 — длительность недели в часах, а значение параметра K вам предстоит подобрать самостоятельно (в зависимости от длины подготовленного ряда, можно для начала взять K равным 2-5).

Если вы собрали данные за несколько лет, аналогичные признаки можно использовать для учёта годовой сезонности. Длина года — 8766 часов.

Если в данных есть линейный или описываемый ещё какой-то простой функцией $f$ тренд, стоит добавить к регрессионным признакам вектор $[1,\dots,T]$ или, соответственно, $f([1,\dots,T])$.

```python
# для фильтрации по району
# возращает границы по номеру региона
def region_edges(i: int) -> list:
    m = 50
    # NY edges
    w = -74.25559
    e = -73.70001
    n = 40.91553
    s = 40.49612
    x_step = (e-w)/m
    y_step = (n-s)/m
    x = np.ceil(i/m)-1
    y = (i-1) % m
    return [(w+x*x_step, s+y*y_step),
            (w+x*x_step, s+y*y_step+y_step),
            (w+x*x_step+x_step, s+y*y_step+y_step),
            (w+x*x_step+x_step, s+y*y_step),
            ]
```

## регрессионные признаки

Пусть считается, что значение ряда y в момент времени t задаётся следующим образом:
$$y_t =\sum\limits_{j=1}^k β_jx_{jt} + z_t$$
то есть фактически y задаётся линейной регрессией, но остатки этой регрессии прогнозируются при помощи
модели ARIMA

В качестве регресссионых признаков берутся тренд и различные сезонные признаки

### тренд

тренд вида: $$trend_t = a_0+a_1*t+a_2t^2+...$$

### сезонные признаки

#### гармоники по длинным периодам сезонности (неделя T = 168 часов)

```python
t = np.array(y.index.map(mapping)).reshape(-1,1)
```

```python
cos_features_w = []
sin_features_w = []
T = 7*24
Kw=7*4 # дней (4 периодов в сутках по 6 часов)
for k in range(1, Kw+1):
    arg = t*k*2*np.pi/T
    cos_features_w.append(np.cos(arg)[:, 0])
    sin_features_w.append(np.sin(arg)[:, 0])

print(np.array(cos_features_w).T.shape,np.array(sin_features_w).T.shape)
```

#### гармоники по длинным периодам сезонности (квартал T = 8766/4 часов)

```python
cos_features_q = []
sin_features_q = []
T = 8766/4 
Kq= 13 # 13 недель
for k in range(1, Kq+1):
    arg = t*k*2*np.pi/T
    cos_features_q.append(np.cos(arg)[:, 0])
    sin_features_q.append(np.sin(arg)[:, 0])

print(np.array(cos_features_q).T.shape,np.array(sin_features_q).T.shape)
```

#### гармоники по длинным периодам сезонности (год T = 8766 часов)

```python
cos_features_y = []
sin_features_y = []
T = 24*365.25 
Ky= 13 # месяцев
for k in range(1, Ky+1):
    arg = t*k*2*np.pi/T
    cos_features_y.append(np.cos(arg)[:, 0])
    sin_features_y.append(np.sin(arg)[:, 0])

print(np.array(cos_features_y).T.shape,np.array(sin_features_y).T.shape)
```

### объединение признаков

```python
features = {"week_fourier": np.c_[np.array(cos_features_w).T,
                                  np.array(sin_features_w).T],
#             "month_fourier": np.c_[np.array(cos_features_m).T,
#                                   np.array(sin_features_m).T],
            "quarter_fourier": np.c_[np.array(cos_features_q).T,
                                     np.array(sin_features_q).T],
            "year_fourier": np.c_[np.array(cos_features_y).T,
                                  np.array(sin_features_y).T],
            "trend": t_features[:, 1:],
            "dayofweek": dayofweek_features,
            "month": month_features,
            "quarter": quarter_features,
            "year": year_features,
            "holidays": holiday_name_features,
            "pre_holidays": pre_holiday_name_features,
            "post_holidays": post_holiday_name_features,
            "holiday_ind": holidays_ind,
            }

# train data
X_train = np.concatenate(list(features.values()), axis=1)

# for coeff searching
columns_bin_width = [f.shape[1] for f in list(features.values())]
columns_bin_edges = [-1]
for column_width in columns_bin_width:
    columns_bin_edges.append(columns_bin_edges[-1]+column_width)
# columns_bin_edges[0] = 0
columns_map = {}
for i, k in enumerate(features.keys()):
    columns_map[k] = slice(columns_bin_edges[i]+1, columns_bin_edges[i+1]+1,1)


columns_map
```

```python
# линейна модель (Lasso для зануления коллинеарных и малоинформативных признаков)
from sklearn.linear_model import ElasticNet
target = y_boxcox
reg_model = ElasticNet(random_state=0,max_iter = 1000, l1_ratio=0.1, fit_intercept=True).fit(X_train, target)
predictions = reg_model.predict(X_train)
```

#### тренд

```python
%pylab inline

f_value = 0
for fname  in ["trend", 
               "month", 
               "quarter", 
               "year", 
               "holidays", 
               "pre_holidays", 
               "post_holidays", 
               "holiday_ind"
              ]:
    coefs = reg_model.coef_[columns_map[fname]].reshape(1, -1)
    f_value += coefs.dot(features[fname].T)
    print(fname+":\n",coefs)
```

```python
plt.figure(figsize(20, 10))
# последние n недели
n = 15
x = pd.to_datetime(y.index)
# целевой признак
plt.plot(x[-n*168:], target[-n*168:], color="orange")
plt.plot(x[-n*168:], f_value[0][-n*168:]+reg_model.intercept_, color="black")
plt.show()
```

# Дифференцирование

```python
def diff(x, k):
    """Дифференцирование k порядка"""
    return x - x.shift(k)
```

```python
# trend_diff_year = diff(pd.Series(remains, index=y.index), 8766)
# trend_diff_quarter = diff(trend_diff_year, 24*7*13)
# trend_diff_week = diff(pd.Series(remains, index=y.index), 24*7)
# дневное дифф
trend_diff_day = diff(pd.Series(pd.Series(remains, index=y.index), index=y.index), 24)
# почасовое дифф
trend_diff_hour = diff(trend_diff_day, 1)
```

```python
plt.figure(figsize(20,10))

data_stl = sm.tsa.seasonal_decompose(trend_diff_hour.dropna(), 
                                     period=24)
n = 52*2
data_stl.plot()
plt.show()
plt.plot(data_stl.resid[-n*168:], c = "orange")
plt.plot(data_stl.trend[-n*168:], c = "red")
plt.plot(data_stl.seasonal[-n*168:], c = "black")
plt.show()
```

```python
print("Критерий Дики-Фуллера: p=%f" % adfuller_test(trend_diff_hour.dropna()))
```

# Поиск лучшей модели

**Выберите оптимальную по AIC модель; постройте график исходного ряда и нарисуйте поверх предсказания выбранной модели. Если модель плохо описывает данные, попробуйте увеличить количество синусов и косинусов KKK или поменять порядки дифференцирования.**

```python
from itertools import product
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
print(len(parameters_list))
parameters_list[:10]
```

```python
from tqdm.notebook import tqdm_notebook
import warnings
```

**выбрал модель (1, 1, 6, 6)**

```python
# модель (1, 1, 6, 6)
parameters_list = [(1, 1, 6, 6)]
parameters_list
```

```python
%%time
n=13 # quartal
S = 24*7 
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in tqdm_notebook(parameters_list):
    # try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model = sm.tsa.statespace.SARIMAX(endog=remains[-n*S:],
#                                           exog=predictions,
                                          order=(param[0], d, param[1]),
                                          seasonal_order=(param[2], D, param[3], 24)).fit(disp=True, low_memory=True)
    # выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    # сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    print([param, model.aic])

warnings.filterwarnings('default')

print(sorted(results, key= lambda x: x[1])[:5])
print(best_model.summary())
```

# Визуальное отображение модели

```python
def invboxcox(y, lmbda):
    """Обратная функция преобразования Бокса-Кокса"""
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
# результаты на 13 неделях
n = 6
time_rane = slice(-n*S, None)
# values
plt.plot(pd.Series(y["count"][time_rane], index = y[time_rane].index ), color = "red")
# predictions
plt.plot(invboxcox(pd.Series(predictions[time_rane], index = y[time_rane].index )+best_model.fittedvalues[time_rane],lmbda), color = "black")
plt.title("Values and predictions")
plt.show()
```
